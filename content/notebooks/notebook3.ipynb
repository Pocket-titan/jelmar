# Stats


```python
from scipy.stats import norm
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

sns.set_theme(style="ticks", palette="Set2", rc={"axes.grid": True})

pdf_x = lambda x: norm.pdf(x)
pdf_y = lambda y: norm.pdf(y, scale=2)

values = np.linspace(-5, 5, num=1000)
pdf_z = np.vectorize(
    lambda z: np.trapz(pdf_x(values) * pdf_y(z / values) * 1 / np.abs(values), values)
)

x = pdf_x(values)
y = pdf_y(values)
z = pdf_z(values)

plt.plot(values, x, label="$X$")
plt.plot(values, y, label="$Y$")
plt.plot(values, z, label="$Z$")
plt.legend()
```
